### **✍️ 최댓값 찾기**

image.png

```
input = [3, 5, 6, 1, 2, 4]


def find_max_num(array):
    max_num = array[0]
    for num in array:
        if num > max_num:
            max_num = num
    return max_num

result = find_max_num(input)
print(result)
```

### **✍️ 최빈값 찾기**

image.png

```
def find_alphabet_occurrence_array(string):
    alphabet_occurrence_array = [0] * 26

    for char in string:
        if char.isalpha():
            order = ord(char) - ord("a")
            alphabet_occurrence_array[order] += 1

    return alphabet_occurrence_array


print(find_alphabet_occurrence_array("hello my name is sparta"))
```

```
input = "hello my name is sparta"


def find_max_occurred_alphabet(string):
    alphabet_occurrence_array = [0] * 26

    for char in string:
        if char.isalpha():
            array_idx = ord(char) - ord("a")
            alphabet_occurrence_array[array_idx] += 1

    max_cnt = 0
    index = 0
    for idx in range(len(alphabet_occurrence_array)):
        cnt = alphabet_occurrence_array[idx]
        if cnt > max_cnt:
            max_cnt = cnt
            index = idx
    
    return chr(index+ord("a"))


result = find_max_occurred_alphabet(input)
print(result)
```

### **✍️ 곱하기 or 더하기**

image.png

```
input = [0, 3, 5, 6, 1, 2, 4]


def find_max_plus_or_multiply(array):
    output = 0
    for num in array:
        # output이 0 혹은 1 (i.e. 1이하)이거나, num이 0혹은 1일 때 +
        if output <= 1 or num <= 1:
            output += num
        else:
            output *= num
    return output


result = find_max_plus_or_multiply(input)
print(result)
```

### **✍️ 반복되지 않는 문자**

image.png

```
input = "abadabac"


def find_not_repeating_character(string):
    alphabet_array = [0] * 26
    # print(alphabet_array)
    for char in string:
        if char.isalpha():
            array_index = ord(char) - ord("a")
            alphabet_array[array_index] += 1
    
    for char in string:
        if char.isalpha():
            array_index = ord(char) - ord("a")
            if alphabet_array[array_index] == 1:
                return char
                break
    return "_"


result = find_not_repeating_character(input)
print(result)
```

### **✍️ 소수 나열하기**

image.png

```
input = 20


def find_prime_list_under_number(number):
    prime_list = [] # by definition

    for num in range(2, input+1):
        for prime in prime_list:
            if num % prime == 0 and prime * prime <= num:
                break
        else:
            prime_list.append(num)
                
    return prime_list


result = find_prime_list_under_number(input)
print(result)
```

### **✍️ 문자열 뒤집기**

image.png

```
input = "011110"


def find_count_to_turn_out_to_all_zero_or_all_one(string):
    all_zero_cnt = 0
    all_one_cnt = 0

    # check 0th index of the input string
    if string[0] == "0":
        all_one_cnt += 1
    elif string[0] == "1":
        all_zero_cnt += 1

    # count # of required changes for each of 0>1 and 1>0 case
    for idx in range(len(string)-1):
        if string[idx] != string[idx+1]:
            if string[idx+1] == "0":
                all_one_cnt += 1
            elif string[idx+1] == "1":
                all_zero_cnt += 1

    # find the minimum count, comparing 0>1 and 1>0 case
    if all_zero_cnt > all_one_cnt:
        return all_one_cnt
    elif all_zero_cnt < all_one_cnt:
        return all_zero_cnt
    else:
        return all_zero_cnt


result = find_count_to_turn_out_to_all_zero_or_all_one(input)
print(result)
```
